```python
# Import dependencies
import pandas as pd
import os
```

```python
# Read file into DataFrame
coffee_shop_file = os.path.join("../", "Resources", "ratings_and_sentiments.csv")
coffee_shop_df = pd.read_csv(coffee_shop_file)
coffee_shop_df
```

```python
# Find all reviews with a number rating greater than or equal to 4.
coffee_shop_df[coffee_shop_df["num_rating"]>= 4]
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.
overall_count = coffee_shop_df[coffee_shop_df["overall_sent"]>= 4]
```

```python
# Return the count of the above query.
len(overall_count)
#overall_count.count()
```

```python
# Change the name of the review_text column to review
renamed_df = coffee_shop_df.rename(columns ={"review_text": "review"})
renamed_df.head()
```

```python
# Change the name of the bool_HIGH column to bool_high
renamed_df = renamed_df.rename(columns ={"bool_HIGH": "bool_high"})
renamed_df.head()
```

```python
# Save your changes to the file as a pipe delimited file.
renamed_df.to_csv("new_file.csv", sep='|')
```
