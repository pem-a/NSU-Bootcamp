```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a pandas df 
ride_count = pd.read_csv('../Resources/pyber_ridercount.csv')
ride_count
```

```python
# Create a figure and axes object 
fig, ax = plt.subplots()
```

```python
# Add a bar for the rural column mean/std
ax.bar('Rural', 
       ride_count['Rural'].mean(), 
       yerr=ride_count['Rural'].std(), 
       align='center',
       alpha=0.5,
       capsize=10)
```

```python
# Add a bar for the urban column mean/std
ax.bar('Urban', 
       ride_count['Urban'].mean(), 
       yerr=ride_count['Urban'].std(), 
       align='center',
       alpha=0.5,
       capsize=10)
```

```python
# Add a bar for the suburban column mean/std
ax.bar('Suburban', 
       ride_count['Suburban'].mean(), 
       yerr=ride_count['Suburban'].std(), 
       align='center',
       alpha=0.5,
       capsize=10)
```

```python
# Set the y label and add grid for y axis
ax.set_ylabel('Ride Average')
ax.yaxis.grid(True)
```

```python
plt.show()
```
