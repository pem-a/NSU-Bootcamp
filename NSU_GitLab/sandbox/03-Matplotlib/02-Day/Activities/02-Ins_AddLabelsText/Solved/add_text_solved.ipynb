### Add Text in The Plot - Solved Version

```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Add simple text
plt.text(0.2, 0.2, 'Text goes here')
plt.show()
```

```python
# Add text w/ font sixe and color
plt.text(0.4, 0.6, 'This look better', fontsize=15, color='red')
plt.show()
```

```python
# Add text right alignment
plt.text(0.9, 0.4, 
         'Indent to the right', 
         fontsize=20, 
         color='blue', 
         horizontalalignment='right')
plt.show()
```
