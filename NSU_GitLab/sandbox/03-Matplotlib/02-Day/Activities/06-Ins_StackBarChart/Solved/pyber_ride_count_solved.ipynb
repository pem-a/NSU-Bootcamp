```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset into a pandas df and make sure to set index to column 0
ride_count = pd.read_csv('../Resources/pyber_ridercount.csv', index_col=0)
ride_count
```

```python
# Create a figure and axes object 
fig, ax = plt.subplots()
```

```python
# Add bars for "Urban" with the label "Urban"
ax.bar(ride_count.index, ride_count["Urban"], label="Urban")
```

```python
# Stack bars for "Suburban" on top with label "Suburban"
ax.bar(ride_count.index, ride_count["Suburban"], bottom=ride_count["Urban"], label="Suburban")
```

```python
# Stack bars for "Rural" on top of that with label "Rural"
ax.bar(ride_count.index, ride_count["Rural"], bottom=ride_count["Urban"] + ride_count["Suburban"], label='Rural')
```

```python
# Rotataing the tick labels vertically
ax.set_xticklabels(ride_count.index, rotation=90)
```

```python
# Setting the y label 
ax.set_ylabel('Number of Rides')
```

```python
# Display the legend
ax.legend()
```

```python
# Use plt.tight_layout() to improve the spacing in the figure
plt.tight_layout()
plt.show()
```
