```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset into a pandas df and make sure to set index to column 0
minefields = pd.read_csv('../Resources/sa_minefields.csv', index_col='countries')
minefields
```

```python
# Create a figure and axes object with figure size 8 inches by 10 inches
fig, ax = plt.subplots(figsize=(8,10))
```

```python
# Add bars for "iron ore" with the label "Iron Ore"
ax.bar(minefields.index, minefields['iron_ore'], label="Iron Ore")
```

```python
# Stack bars for "copper" on top with label "Copper"
ax.bar(minefields.index, minefields['copper'], bottom=minefields['iron_ore'], label='Copper')
```

```python
# Stack bars for "gold" on top of that with label "Gold"
ax.bar(minefields.index, minefields['gold'], bottom=minefields['iron_ore'] + minefields['copper'], label='Gold')
```

```python
# Rotating the tick labels vertically
ax.set_xticklabels(minefields.index, rotation=90)
```

```python
# Setting the y label 
ax.set_ylabel('Number of Minefield')
```

```python
# Display the legend
ax.legend()
```

```python
# Use plt.tight_layout() to improve the spacing in the figure
plt.tight_layout()
plt.show()
```
