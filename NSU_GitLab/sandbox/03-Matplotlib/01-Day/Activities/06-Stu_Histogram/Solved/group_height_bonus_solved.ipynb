### Histogram Group Height Bonus - Solved Version

* In this activity, you will plot a histogram to show a simulated dataset containing the height distribution in a group of 100 people.

* **Instructions:** 

  * Import dependencies.

  * Create an array using NumPy's `np.random.normal()` function, passing the following as arguments:

    * The mean height of the group should be 175 cm.

    * The standard deviation should be 4 cm. 

    * The total number of people should be 100.

  * Finally build the histogram using `plt.hist()`

  * **Bonus:** Create labels for the axes and Grid to the plot.


```python
%matplotlib notebook
```

```python
# Import dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Add variable for mean and std and create a numpy array using numpy
mu = 175
sigma = 4
array = np.random.normal(mu, sigma, size=100)
```

```python
# Create histogram
plt.hist(array, 10)
```

### BONUS: Create Labels for Axes and Add Grid

```python
# Add labels and axes
plt.xlabel('Height')
plt.ylabel('# of People')
plt.grid()
plt.show()
```
