```python
# Import dependencies
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Create an array representing the twelve months of the year
months = np.arange(1, 13, 1)
months
```

```python
# Create a second array with the average temperature per month in Farenheit
avg_temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert F to C
avg_temp_c = [(x-32) * .56 for x in avg_temp_f]
avg_temp_c
```

```python
# Plot both F and C in the same chart
plt.plot(months, avg_temp_f)
plt.plot(months, avg_temp_c)
```
