### Histogram - Solved Version

```python
%matplotlib inline
```

```python
# Import dependencies
import numpy as np
import matplotlib.pyplot as plt  
```

```python
# Create a random array
array_one = [94, 41, 68, 10, 48, 31, 73, 61, 24, 31]
```

```python
# Create a histogram
plt.hist(array_one)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a histogram passing a bins argument as 20 
plt.hist(array_one, bins=20)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a second array using numpy and specify mu and sigma
mu = 160
sigma = 14
array_two = np.random.normal(mu, sigma, size=400)
```

```python
# Create histogram
plt.hist(array_two, 20)
plt.xlabel('Bin Range')
plt.ylabel('Frequency')
plt.show()
```
