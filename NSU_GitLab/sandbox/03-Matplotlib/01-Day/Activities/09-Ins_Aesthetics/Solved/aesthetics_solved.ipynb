```python
# Import dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate the x axis with integers starting at 0 and ending at 10 with 0.1 steps
x_axis = np.arange(0, 10, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
# Add a semi-transparent horizontal line at axis y = 0
plt.hlines(0, 0, 10, alpha=0.25)
```

```python
# Using circles and triangle up as markers setting the color blue and red respectively
plt.plot(x_axis, sin, linewidth=0, marker='o', color='blue')
plt.plot(x_axis, cos, linewidth=0, marker='^', color='red')
```

```python
# Add labels to the x and y axes
plt.title("Juxtaposed Sine and Cosine Curves")
plt.xlabel("Input (Sampled Real Numbers from 0 to 10)")
plt.ylabel("Value of Sine (blue) and Cosine (red)")
```

```python
# Set x and y axes limits
plt.xlim(0, 10)
plt.ylim(-1, 1)
```

```python
# Set a grid to the plot
plt.grid()
```

```python
# Save the plot and display it
plt.savefig("../Images/sin_cos_w_markers.png")
plt.show()
```
