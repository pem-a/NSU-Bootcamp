```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import the data into pandas from CSV
citibike_df = pd.read_csv('../Resources/JC-202010-citibike-tripdata.csv')
citibike_df.head()
```

```python
# Split up our data into groups based on user type
user_type = citibike_df.groupby('usertype')
```

```python
# Find out how many bike trips each customer type took
user_type_count = user_type['usertype'].count()
user_type_count
```

```python
# Create a bar chart based on user type count
user_type_count_chart = user_type_count.plot(kind='bar')

# Set the xlabel and ylabel using class methods
user_type_count_chart.set_xlabel("Customer Type")
user_type_count_chart.set_ylabel("Number of Trips")

# Show plot and make it plot size to fit outcome cell

plt.tight_layout()
plt.grid(True)
plt.show()
```
