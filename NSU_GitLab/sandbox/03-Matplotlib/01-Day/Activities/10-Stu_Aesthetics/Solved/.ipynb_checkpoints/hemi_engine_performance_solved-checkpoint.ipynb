```python
# Import dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the X and Y axis
x_axis = np.arange(500, 7500, 1000)
bhp = [100, 200, 300, 400, 500, 600, 700]
lbft = [400, 500, 600, 650, 625, 600, 550]
```

```python
# Plot the chart and apply some styling
engine_power, = plt.plot(x_axis, bhp, marker="D", color="red", label="Engine Power")
torque, = plt.plot(x_axis, lbft, marker="o", color="blue", label="Torque")
```

```python
# Add labels to x and y axes and add a title
plt.title('6.2L HEMI HELLCAT HORSEPOWER/TORQUE')
plt.xlabel('RPM')
plt.ylabel('Power(bhp), Torque(lbft)')
```

```python
# Set the x and y limits
plt.xlim(200, 7200)
```

```python
plt.ylim(0, 800)
```

```python
# Add legend to the chart
plt.legend(handles=[engine_power, torque], loc="lower right")
```

```python
# Set a grid to the plot
plt.grid(True)
```

```python
# Save plot
plt.savefig('../Images/hemi_engine_performance.png')
```
