```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Get the file path for running team one
team1_file = os.path.join("../", "Resources","team-1.csv")
team2_file = os.path.join("../", "Resources","team-2.csv")
team1_df =pd.read_csv(team1_file)
team2_df =pd.read_csv(team2_file)
```

```python
# Read the CSV file into a DataFrame
team1_df
```

```python
team2_df
```

```python
# Melt the DataFrame
team1_df.melt()
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
runner_name_df = team1_df.melt(id_vars = "runner_name")
runner_name_df
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
coach_name_df = team1_df.melt(id_vars = "coach_name")
coach_name_df
```

```python
# Renaming the columns
renamed_df = team1_df.melt(id_vars =["runner_name","coach_name"],var_name ="run_num",value_name ="time")
renamed_df
```

```python
# Remove coach name from DataFrame
renamed_df = renamed_df.drop("coach_name",axis=1)
renamed_df
```

```python
# Show final result
renamed_df.melt()
```

```python
# Save results to a file
renamed_df.to_csv("new_file.csv")
```
